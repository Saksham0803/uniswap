```python
import matplotlib.pyplot as plt
import random
```

```python
x=[]
for i in range(1,31):
    x.append(str(i)+'/11')
for i in range(1,32):
    x.append(str(i)+'/12')
for i in range(1,32):
    x.append(str(i)+'/1')
print(x)
y=[]
a=183578
y.append(a)
for i in range((len(x)-1)):
    b=random.randint(-10,20)
    y.append(a+int(b*2344))
print(y)
print(len(x),len(y))
```

```python
plt.style.use('seaborn')
plt.xlabel('date')
plt.ylabel("VolumeToken0")
plt.plot(x,y)
plt.tight_layout()
plt.gcf().autofmt_xdate()
```
